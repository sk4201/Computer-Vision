```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```
image1=cv2.imread('image1.jpeg')
image2=cv2.imread('image2.jpeg')
image3=cv2.imread('image3.jpeg')
image=image1
```

```
#resizing image2 to become equal to image1
image1 = cv2.resize(image1,(290,160), interpolation = cv2.INTER_AREA) 
image2 = cv2.resize(image2,(290,160), interpolation = cv2.INTER_AREA)

```

```
image1.shape
```

```

cv2_imshow(image1)
```

```
#Now cropping image1 and image2
crop_img1=image1[0:80,0:150]
crop_img2=image2[0:80,0:150]
```

```
cv2_imshow(crop_img1)
```

```
cv2_imshow(crop_img2)
```

```
crop_img1[35:50,50:90]=image2[35:50,50:90]
crop_img2[35:50,50:90]=image[35:50,50:90]
```

```
cv2_imshow(crop_img1)
```
